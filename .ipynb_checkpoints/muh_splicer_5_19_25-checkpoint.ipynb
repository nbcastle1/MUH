```python
class ProcessedTrialData:
    def __init__(self, metadata_path, data_root_dir, processed_file='processed_data.pkl', 
                 force_reprocess=False, min_complete_strides=20, debug=True):
        self.metadata_path = metadata_path
        self.data_root_dir = data_root_dir
        self.processed_file = processed_file
        self.min_strides = min_complete_strides
        self.metadata = None
        self.processed_data = {}
        self.debug = debug
        
        # Create figures directory if it doesn't exist
        self.figures_dir = Path('figures_tst')
        os.makedirs(self.figures_dir, exist_ok=True)
        
        # Define trial type mappings
        self.trial_type_mapping = {
            'primer': 'vis1',
            'trial': 'invis',
            'vis': 'vis2',
            'pref': 'pref'  # keep pref as is
        }       
        
        # Patterns to identify trial metadata from filenames
        self.trial_pattern = re.compile(
            r'(?P<subject_id>[A-Za-z0-9]+)_'  # Subject ID
            r'(?P<trial_type>vis1|vis2|invis|pref)_'  # Trial type
            r'(?P<session>\d{8})_'  # Session date (YYYYMMDD)
            r'(?P<trial_num>\d+)'  # Trial number
            r'(?P<fragment>_frag\d+)?'  # Optional fragment indicator
            r'\.csv'  # File extension
        )
        
        # Load or process data
        if not force_reprocess and os.path.exists(self.processed_file):
            self._load_processed_data()
        else:
            self._load_all_data()
            self._save_processed_data()

    def _find_trial_files(self, subject_dir: Path) -> Dict[str, List[Path]]:
        """Group related trial files for a subject based on metadata"""
        trial_groups = {}
        
        for file in subject_dir.glob('*.csv'):
            match = self.trial_pattern.match(file.name)
            if not match:
                continue
                
            # Create grouping key (excluding fragment info)
            base_name = (
                f"{match.group('subject_id')}_"
                f"{match.group('trial_type')}_"
                f"{match.group('session')}_"
                f"{match.group('trial_num')}"
            )
            
            if base_name not in trial_groups:
                trial_groups[base_name] = []
            trial_groups[base_name].append(file)
        
        return trial_groups
    
    def _load_trial_file(self, file_path: Path) -> Optional[pd.DataFrame]:
        """Load and validate a single trial file"""
        try:
            df = pd.read_csv(file_path)
            
            # Basic validation
            required_cols = {'Stride Number', 'Success', 'Constant', 'Target size'}
            if not required_cols.issubset(df.columns):
                return None
                
            # Ensure stride numbers are numeric and unique
            df['Stride Number'] = pd.to_numeric(df['Stride Number'], errors='coerce')
            df = df[df['Stride Number'].notna()]
            df = df.drop_duplicates(subset=['Stride Number'])
            
            if df.empty:
                return None
                
            return df
            
        except Exception as e:
            print(f"Error loading {file_path}: {str(e)}")
            return None
    
    def _find_best_continuation(self, current_df: pd.DataFrame, candidate_dfs: List[pd.DataFrame]) -> Optional[Tuple[pd.DataFrame, int]]:
        """
        Find the best matching continuation among candidate fragments
        
        Args:
            current_df: The base trial data
            candidate_dfs: Potential continuation fragments
            
        Returns:
            Tuple of (best_match_df, overlap_count) or None
        """
        best_match = None
        best_overlap = 0
        last_stride = current_df['Stride Number'].max()
        
        for candidate in candidate_dfs:
            # Get strides that would continue our current data
            continuation = candidate[candidate['Stride Number'] > last_stride]
            
            if continuation.empty:
                continue
                
            # Check how well this continues the sequence
            expected_next = last_stride + 1
            first_new_stride = continuation['Stride Number'].min()
            
            # Perfect continuation gets priority
            if first_new_stride == expected_next:
                return (candidate, 0)
                
            # Otherwise look for highest overlap
            overlap = len(continuation[continuation['Stride Number'].isin(
                range(expected_next, expected_next + 10)
            )])
            
            if overlap > best_overlap:
                best_match = candidate
                best_overlap = overlap
        
        return (best_match, best_overlap) if best_match else None
    
    def _splice_trial(self, file_group: List[Path]) -> Optional[pd.DataFrame]:
        """
        Splice together fragments of a single trial and print status messages.
        
        Args:
            file_group: List of related trial files
            
        Returns:
            Combined DataFrame if valid, None otherwise
        """
        if len(file_group) == 1:
            # Single file case
            df = self._load_trial_file(file_group[0])
            if df is not None:
                print(f"‚úì Using single file: {file_group[0].name} ({len(df)} strides)")
            return df if df is not None and len(df) >= self.min_strides else None
        
        # Multi-file case
        print(f"\nüîç Found {len(file_group)} fragments for trial: {file_group[0].stem.split('_frag')[0]}")
        fragments = []
        for i, file in enumerate(file_group, 1):
            df = self._load_trial_file(file)
            if df is not None:
                fragments.append(df)
                print(f"  - Fragment {i}: {file.name} ({len(df)} strides)")
        
        if not fragments:
            print("  ‚úó All fragments failed validation")
            return None
            
        fragments.sort(key=lambda x: x['Stride Number'].min())
        
        # Start combining
        combined = fragments[0].copy()
        remaining_fragments = fragments[1:]
        print(f"  ‚Ä¢ Base fragment: {file_group[0].name} (Strides {combined['Stride Number'].min()}-{combined['Stride Number'].max()})")
        
        while remaining_fragments:
            result = self._find_best_continuation(combined, remaining_fragments)
            if not result:
                break
                
            next_df, overlap = result
            remaining_fragments.remove(next_df)
            
            last_stride = combined['Stride Number'].max()
            new_data = next_df[next_df['Stride Number'] > last_stride]
            
            if not new_data.empty:
                combined = pd.concat([combined, new_data], ignore_index=True)
                print(f"  ‚ûî Added continuation: Strides {new_data['Stride Number'].min()}-{new_data['Stride Number'].max()} "
                      f"(Overlap quality: {overlap}/10)")
        
        # Validation messages
        if len(combined) < self.min_strides:
            print(f"  ‚úó Combined trial too short ({len(combined)} < {self.min_strides} strides)")
            return None
            
        stride_nums = sorted(combined['Stride Number'].unique())
        expected = list(range(min(stride_nums), max(stride_nums) + 1))
        
        if len(stride_nums) != len(expected) or any(s != e for s, e in zip(stride_nums, expected)):
            print(f"  ‚úó Discontinuous strides in combined trial (Gaps detected)")
            return None
        
        print(f"  ‚úì Successfully combined: {len(combined)} continuous strides")
        return combined

    def _process_trial_data(self, df, trial_type):
        """Process trial data to calculate target size, one entry per unique stride"""
        if df is None or df.empty:
            return None, {}  # Return empty anomaly dict for consistency
        
        anomaly_dict = {}
        
        # Skip processing for pref trials
        if trial_type == 'pref':
            df = df.drop_duplicates(subset='Left heel strike', keep='last')
            return df, anomaly_dict
        
        required_cols = ['Stride Number', 'Success', 'Upper bound success', 
                         'Lower bound success', 'Constant']
        missing_cols = [col for col in required_cols if col not in df.columns]
        if missing_cols:
            print(f"Missing required columns in {trial_type}: {missing_cols}")
            return None, {}
        
        try:
            df = df.sort_values('Stride Number')
            df['Target size'] = df['Upper bound success'] - df['Lower bound success']
            df = df.drop_duplicates(subset='Stride Number', keep='last')
            if 'Sum of gains and steps' in df.columns:
                df['Sum of gains and steps'] = 1.5*df['Sum of gains and steps']
            df, anomaly_dict = self._detect_anomalies(df)
            return df, anomaly_dict
        
        except Exception as e:
            print(f"Error processing {trial_type} data: {str(e)}")
            return None, {}

    def _load_subject_data(self, subject_id):
        subject_dir = Path(self.data_root_dir) / subject_id
        if not subject_dir.exists():
            print(f"‚ùå Directory not found: {subject_dir}") if self.debug else None
            return None
    
        trial_data = {}
        original_trial_types = ['primer', 'trial', 'vis', 'pref']
        
        for original_type in original_trial_types:
            try:
                files = self._get_trial_files(subject_dir, original_type)
                if not files:
                    continue
                    
                if len(files) > 1:
                    # Combine multiple fragments
                    combined_df = self._combine_txt_fragments(files)
                    if combined_df is not None:
                        print(f"  ‚ûî Combined {len(files)} {original_type} fragments") if self.debug else None
                        new_type = self.trial_type_mapping[original_type]
                        processed_df, anomaly_dict = self._process_trial_data(combined_df, original_type)
                        trial_data[new_type] = {
                            'data': processed_df,
                            'anomalies': anomaly_dict
                        }
                else:
                    # Single file case
                    df = pd.read_csv(files[0], sep='\t')
                    new_type = self.trial_type_mapping[original_type]
                    processed_df, anomaly_dict = self._process_trial_data(df, original_type)
                    trial_data[new_type] = {
                        'data': processed_df,
                        'anomalies': anomaly_dict
                    }
                    
            except Exception as e:
                print(f"Error processing {subject_id}/{original_type}: {str(e)}")
                trial_data[self.trial_type_mapping[original_type]] = None
    
        return trial_data

    def _load_processed_data(self):
        """Load previously processed data from file"""
        with open(self.processed_file, 'rb') as f:
            self.processed_data = pickle.load(f)
            self.metadata = pd.DataFrame.from_dict(
                {subj: data['metadata'] for subj, data in self.processed_data.items()}, 
                orient='index'
            )
            
    def get_trial_df(self, trial_dict):
        """Helper method to extract dataframe from trial dictionary"""
        if trial_dict is None:
            return None
        return trial_dict.get('data')
        
    def _save_processed_data(self):
        """Save processed data to file for future use"""
        with open(self.processed_file, 'wb') as f:
            pickle.dump(self.processed_data, f)

    def _load_metadata(self):
        """Load and preprocess metadata, filtering out invalid entries."""
        self.metadata = pd.read_csv(self.metadata_path)
        self.metadata['DOB'] = pd.to_datetime(self.metadata['DOB'], errors='coerce')
        self.metadata['Session Date'] = pd.to_datetime(self.metadata['Session Date'], errors='coerce')
        self.metadata = self.metadata.dropna(subset=['ID', 'age_months'])

    def _get_trial_files(self, subject_dir, trial_prefix):
        """Find files for a trial type with silent handling for single files"""
        all_files = sorted(subject_dir.glob(f"{trial_prefix}*.txt"))
        
        if not all_files:
            if self.debug:
                print(f"  ‚ö†Ô∏è No files found for {trial_prefix}")
            return None
        
        # Special case for pref trials
        if trial_prefix == 'pref':
            largest_file = max(all_files, key=lambda f: f.stat().st_size)
            if self.debug and len(all_files) > 1:
                print(f"  ‚ö° pref trial - selected largest of {len(all_files)} files:")
                print(f"    ‚Ä¢ {largest_file.name} ({largest_file.stat().st_size/1024:.1f} KB)")
            return [largest_file]
        
        # Process other trial types
        file_info = []
        for f in all_files:
            try:
                with open(f, 'r') as file:
                    header = file.readline().strip().split('\t')
                    stride_col = next((i for i, col in enumerate(header) 
                                     if 'stride' in col.lower() and 
                                     ('num' in col.lower() or 'no' in col.lower())), None)
                    
                    if stride_col is None:
                        continue
                        
                    # Get first and last stride numbers
                    second_line = next(line.strip() for line in file if line.strip())
                    first_stride = float(second_line.split('\t')[stride_col])
                    
                    file.seek(0, 2)
                    file.seek(max(0, file.tell() - 1024), 0)
                    last_line = next((l.strip() for l in reversed(file.readlines()) if l.strip()), None)
                    last_stride = float(last_line.split('\t')[stride_col])
                    
                    file_info.append({
                        'path': f,
                        'first': first_stride,
                        'last': last_stride,
                        'size': f.stat().st_size
                    })
                    
            except Exception:
                continue
        
        # Single file case - return without warnings
        if len(file_info) == 1:
            return [file_info[0]['path']]
        
        # Multi-file processing
        if not file_info:
            return [max(all_files, key=lambda f: f.stat().st_size)]
        
        file_info.sort(key=lambda x: x['first'])
        best_sequence = []
        current_sequence = [file_info[0]]
        
        for file_data in file_info[1:]:
            if file_data['first'] == current_sequence[-1]['last'] + 1:
                current_sequence.append(file_data)
            else:
                if len(current_sequence) > len(best_sequence):
                    best_sequence = current_sequence
                current_sequence = [file_data]
        
        best_sequence = max([best_sequence, current_sequence], key=len)
        
        if len(best_sequence) >= 2:
            if self.debug:
                print(f"  üîó Combined {len(best_sequence)} {trial_prefix} files:")
                for f in best_sequence:
                    print(f"    ‚Ä¢ {f['path'].name} (strides {int(f['first'])}-{int(f['last'])})")
            return [f['path'] for f in best_sequence]
        
        # Silent fallback for non-continuous files
        return [max(file_info, key=lambda x: x['size'])['path']]

    def _load_all_data(self):
        """Load data for all subjects with progress tracking"""
        self._load_metadata()
        total_subjects = len(self.metadata)
        
        if self.debug:
            print(f"üîÑ Processing {total_subjects} subjects...")
        
        for i, (_, row) in enumerate(self.metadata.iterrows(), 1):
            subject_id = row['ID']
            if self.debug:
                print(f"\n[{i}/{total_subjects}] Processing {subject_id}...")
                
            subject_data = self._load_subject_data(subject_id)
            if subject_data:
                self.processed_data[subject_id] = {
                    'metadata': row.to_dict(),
                    'trial_data': subject_data
                }

    def _detect_anomalies(self, df):
        """Flags anomalies and logs reasons."""
        if df is None or df.empty:
            return df, {}

        df = df.copy()
        df['Anomalous'] = False
        anomalies = {}

        # Time-based anomaly
        time_col = next((col for col in ['Time', 'Timestamp', 'Time (s)'] if col in df.columns), None)
        if time_col:
            df[time_col] = pd.to_numeric(df[time_col], errors='coerce')
            time_diff = df[time_col].diff()
            jump_mask = time_diff > time_diff.quantile(0.99) * 5
            for idx in df.index[jump_mask.fillna(False)]:
                df.at[idx, 'Anomalous'] = True
                anomalies.setdefault(idx, []).append('time_jump')

        # Sum of gains and steps anomalies
        if 'Sum of gains and steps' in df.columns:
            high_mask = df['Sum of gains and steps'] > 4
            zero_mask = df['Sum of gains and steps'] == 0

            for idx in df.index[high_mask]:
                df.at[idx, 'Anomalous'] = True
                anomalies.setdefault(idx, []).append('sum_gain_step_high')
            for idx in df.index[zero_mask]:
                df.at[idx, 'Anomalous'] = True
                anomalies.setdefault(idx, []).append('sum_gain_step_zero')

        # Repeated rows
        duplicated_mask = df.duplicated()
        for idx in df.index[duplicated_mask]:
            df.at[idx, 'Anomalous'] = True
            anomalies.setdefault(idx, []).append('duplicate_row')

        return df, anomalies

    def get_anomalies(self, subject_id, trial_type):
        """Get anomaly dictionary for a specific subject and trial type"""
        try:
            return self.processed_data[subject_id]['trial_data'][trial_type]['anomalies']
        except (KeyError, TypeError):
            return None

    def filter_trials(self, max_target_size=None, min_age=None, max_age=None, 
                     required_trial_types=None, min_strides=None, max_strides=None):
        """
        Enhanced filtering that returns a new ProcessedTrialData object with filtered trials
        
        Args:
            max_target_size: Maximum target size to include
            min_age: Minimum age in years to include
            max_age: Maximum age in years to include
            required_trial_types: List of trial types that must be present (e.g., ['invis', 'pref'])
            min_strides: Minimum number of strides required per trial (overrides class default)
            max_strides: Maximum number of strides allowed per trial
            
        Returns:
            A new ProcessedTrialData object with only the filtered trials
        """
        # Create new filtered object
        filtered = ProcessedTrialData(
            metadata_path=self.metadata_path,
            data_root_dir=self.data_root_dir,
            processed_file=self.processed_file,  # Preserve original file
            force_reprocess=False,
            min_complete_strides=min_strides if min_strides is not None else self.min_strides,
            debug=self.debug
        )
        
        # Copy the metadata structure
        filtered.metadata = self.metadata.copy()
        filtered.processed_data = {}
        
        for subject_id, subject_data in self.processed_data.items():
            # Age filtering
            age_years = subject_data['metadata']['age_months'] / 12
            if min_age is not None and age_years < min_age:
                continue
            if max_age is not None and age_years > max_age:
                continue
            
            # Check trial types
            valid_trial_types = True
            if required_trial_types:
                missing_trials = [
                    t for t in required_trial_types 
                    if t not in subject_data['trial_data'] or 
                       subject_data['trial_data'][t] is None or
                       subject_data['trial_data'][t]['data'] is None
                ]
                if missing_trials:
                    continue
            
            # Check target sizes and stride counts
            target_ok = True
            stride_count_ok = True
            subject_trial_data = {}
            
            for trial_type, trial_dict in subject_data['trial_data'].items():
                if trial_dict and trial_dict['data'] is not None:
                    df = trial_dict['data']
                    
                    # Target size check
                    if (max_target_size is not None and 
                        'Target size' in df.columns and 
                        df['Target size'].min() > max_target_size):
                        target_ok = False
                        break
                    
                    # Stride count check
                    n_strides = len(df)
                    if ((min_strides is not None and n_strides < min_strides) or
                        (max_strides is not None and n_strides > max_strides)):
                        stride_count_ok = False
                        break
                    
                    # If passed filters, include this trial
                    subject_trial_data[trial_type] = {
                        'data': df.copy(),
                        'anomalies': trial_dict['anomalies'].copy()
                    }
            
            if target_ok and stride_count_ok and subject_trial_data:
                filtered.processed_data[subject_id] = {
                    'metadata': subject_data['metadata'].copy(),
                    'trial_data': subject_trial_data
                }
        
        return filtered
    
    def print_participant_ages(self):
        """Print ages of all participants in years"""
        if not self.processed_data:
            print("No processed data available.")
            return
        
        print("Participant Ages (in years):")
        for subject_id, data in self.processed_data.items():
            age_months = data['metadata'].get('age_months')
            if age_months is not None:
                age_years = round(age_months / 12, 2)
                print(f"ID: {subject_id}, Age: {age_years} years")
            else:
                print(f"ID: {subject_id}, Age: not available")

    def _combine_txt_fragments(self, file_list):
        """Combine multiple text fragments into one DataFrame"""
        dfs = []
        for f in sorted(file_list):
            try:
                df = pd.read_csv(f, sep='\t')
                dfs.append(df)
                if self.debug:
                    print(f"    ‚úì Loaded {f.name} ({len(df)} rows)")
            except Exception as e:
                print(f"    ‚ùå Failed to load {f.name}: {str(e)}")
        
        if not dfs:
            return None
            
        combined = pd.concat(dfs, ignore_index=True)
        
        # Basic validation
        if 'Stride Number' in combined.columns:
            combined = combined.sort_values('Stride Number')
            combined = combined.drop_duplicates('Stride Number')
        
        return combined
```

```python
class TrialSplicer:
    def __init__(self, data_directory: str, min_complete_strides: int = 20):
        """
        Initialize the TrialSplicer with directory containing trial files
        
        Args:
            data_directory: Path to directory containing trial files
            min_complete_strides: Minimum number of continuous strides to consider a trial valid
        """
        self.data_dir = Path(data_directory)
        self.min_strides = min_complete_strides
        self.file_cache = {}  # Cache loaded files to avoid reloading
        
        # Patterns to identify trial metadata from filenames
        self.trial_pattern = re.compile(
            r'(?P<subject_id>[A-Za-z0-9]+)_'  # Subject ID (e.g., MUH451)
            r'(?P<trial_type>primer|trial|vis|pref)_'  # Your actual trial prefixes
            r'(?P<fragment>\d+)'  # Fragment number
            r'\.txt'  # Your file extension
        )
    
    def find_trial_files(self) -> Dict[str, List[Path]]:
        """Group related trial files based on metadata"""
        trial_groups = {}
        
        for file in self.data_dir.glob('*.csv'):
            match = self.trial_pattern.match(file.name)
            if not match:
                continue
                
            # Create grouping key (excluding fragment info)
            base_name = (
                f"{match.group('subject_id')}_"
                f"{match.group('trial_type')}_"
                f"{match.group('session')}_"
                f"{match.group('trial_num')}"
            )
            
            if base_name not in trial_groups:
                trial_groups[base_name] = []
            trial_groups[base_name].append(file)
        
        return trial_groups
    
    def _load_trial_file(self, file_path: Path) -> Optional[pd.DataFrame]:
        try:
            # Changed to read tab-delimited text files
            df = pd.read_csv(file_path, sep='\t')
            
            # Add validation for required columns
            required_cols = {'Stride Number', 'Success', 'Constant'}
            if not required_cols.issubset(df.columns):
                print(f"‚ö†Ô∏è Missing columns in {file_path.name}")
                return None
                
            # Convert stride numbers to numeric
            df['Stride Number'] = pd.to_numeric(df['Stride Number'], errors='coerce')
            df = df.dropna(subset=['Stride Number'])
            
            print(f"‚úì Loaded {file_path.name} ({len(df)} strides)")
            return df
            
        except Exception as e:
            print(f"‚ùå Failed to load {file_path.name}: {str(e)}")
            return None
    
    def find_best_continuation(self, current_df: pd.DataFrame, candidate_dfs: List[pd.DataFrame]) -> Optional[Tuple[pd.DataFrame, int]]:
        """
        Find the best matching continuation among candidate fragments
        
        Args:
            current_df: The base trial data
            candidate_dfs: Potential continuation fragments
            
        Returns:
            Tuple of (best_match_df, overlap_count) or None
        """
        best_match = None
        best_overlap = 0
        last_stride = current_df['Stride Number'].max()
        
        for candidate in candidate_dfs:
            # Get strides that would continue our current data
            continuation = candidate[candidate['Stride Number'] > last_stride]
            
            if continuation.empty:
                continue
                
            # Check how well this continues the sequence
            expected_next = last_stride + 1
            first_new_stride = continuation['Stride Number'].min()
            
            # Perfect continuation gets priority
            if first_new_stride == expected_next:
                return (candidate, 0)
                
            # Otherwise look for highest overlap
            overlap = len(continuation[continuation['Stride Number'].isin(
                range(expected_next, expected_next + 10)
            )])
            
            if overlap > best_overlap:
                best_match = candidate
                best_overlap = overlap
        
        return (best_match, best_overlap) if best_match else None
    
    def splice_trial(self, file_group: List[Path]) -> Optional[pd.DataFrame]:
        """
        Splice together fragments of a single trial
        
        Args:
            file_group: List of related trial files
            
        Returns:
            Combined DataFrame if valid, None otherwise
        """
        if len(file_group) == 1:
            # Single file - just validate
            df = self.load_trial_file(file_group[0])
            return df if df is not None and len(df) >= self.min_strides else None
        
        # Load all fragments
        fragments = []
        for file in file_group:
            df = self.load_trial_file(file)
            if df is not None:
                fragments.append(df)
        
        if not fragments:
            return None
            
        # Sort fragments by their starting stride number
        fragments.sort(key=lambda x: x['Stride Number'].min())
        
        # Start with earliest fragment
        combined = fragments[0].copy()
        remaining_fragments = fragments[1:]
        
        while remaining_fragments:
            # Find best continuation
            result = self.find_best_continuation(combined, remaining_fragments)
            if not result:
                break
                
            next_df, overlap = result
            remaining_fragments.remove(next_df)
            
            # Get only new strides from next fragment
            last_stride = combined['Stride Number'].max()
            new_data = next_df[next_df['Stride Number'] > last_stride]
            
            if not new_data.empty:
                combined = pd.concat([combined, new_data], ignore_index=True)
        
        # Validate final result
        if len(combined) < self.min_strides:
            return None
            
        # Check stride continuity
        stride_nums = sorted(combined['Stride Number'].unique())
        expected = list(range(min(stride_nums), max(stride_nums) + 1))
        
        if len(stride_nums) != len(expected) or any(s != e for s, e in zip(stride_nums, expected)):
            print(f"Warning: Discontinuous strides in combined trial {file_group[0].stem}")
            return None
            
        return combined
    
    def process_all_trials(self) -> Dict[str, pd.DataFrame]:
        """
        Process all trials in the directory
        
        Returns:
            Dictionary of {trial_name: combined_df}
        """
        trial_files = self.find_trial_files()
        complete_trials = {}
        
        for trial_name, files in tqdm(trial_files.items(), desc="Processing trials"):
            combined = self.splice_trial(files)
            if combined is not None:
                complete_trials[trial_name] = combined
                
        return complete_trials
    
    def save_combined_trials(self, output_dir: str):
        """Save all successfully combined trials to CSV files"""
        output_path = Path(output_dir)
        output_path.mkdir(exist_ok=True)
        
        trials = self.process_all_trials()
        for name, df in trials.items():
            df.to_csv(output_path / f"{name}.csv", index=False)
        
        print(f"Saved {len(trials)} combined trials to {output_dir}")
```

```python
class PerformanceMetricsDB:
    def __init__(self, db_path: str = 'performance_metrics.db'):
        """
        Robust metrics database with duplicate prevention and data integrity checks
        
        Args:
            db_path: Path to database file (default: 'performance_metrics.db')
        """
        self.db_path = Path(db_path)
        self.metrics_df = self._initialize_empty_db()  # Initialize the DataFrame
        
        if self.db_path.exists():
            self._load_db()  # This will populate metrics_df if db exists
            self._clean_db()  # Clean existing data on load
        else:
            # Ensure metrics_df exists even if no db file
            self.metrics_df = self._initialize_empty_db()

    def _initialize_empty_db(self) -> pd.DataFrame:
        """Initialize with all expected columns and proper dtypes"""
        columns=[
            'ID', 'age_years', 'session_date',
            # Trial metrics
            'vis1_sr_max_const', 'vis1_sr_min_const',
            'vis1_sd_max_const', 'vis1_sd_min_const',
            'vis1_msl_max_const', 'vis1_msl_min_const',
            'vis1_error_max_const', 'vis1_error_min_const',
            'vis1_asymmetry_max_const', 'vis1_asymmetry_min_const',
            'vis1_strides_between_success_max_const', 'vis1_strides_between_success_min_const',
            'invis_sr_max_const', 'invis_sr_min_const',
            'invis_sd_max_const', 'invis_sd_min_const',
            'invis_msl_max_const', 'invis_msl_min_const',
            'invis_error_max_const', 'invis_error_min_const',
            'invis_asymmetry_max_const', 'invis_asymmetry_min_const',
            'invis_strides_between_success_max_const', 'invis_strides_between_success_min_const',
            'vis2_sr_max_const', 'vis2_sr_min_const',
            'vis2_sd_max_const', 'vis2_sd_min_const',
            'vis2_msl_max_const', 'vis2_msl_min_const',
            'vis2_error_max_const', 'vis2_error_min_const',
            'vis2_asymmetry_max_const', 'vis2_asymmetry_min_const',
            'vis2_strides_between_success_max_const', 'vis2_strides_between_success_min_const',
            # Preference trial metrics
            'mot_noise', 'pref_asymmetry',
            # Target size info
            'vis1_min_target_size', 'invis_min_target_size', 'vis2_min_target_size',
            # Constant info
            'vis1_max_constant', 'vis1_min_constant',
            'invis_max_constant', 'invis_min_constant',
            'vis2_max_constant', 'vis2_min_constant',
            # Order info
            'invis_max_first', 'invis_min_first'
        ]
        return pd.DataFrame(columns=columns)

    def _load_db(self) -> None:
        """Alternative implementation using combine_first"""
        try:
            with open(self.db_path, 'rb') as f:
                loaded_data = pickle.load(f)
                    
            if isinstance(loaded_data, pd.DataFrame):
                # Create empty DataFrame with our schema
                empty_df = self._initialize_empty_db()
                    
                # Safe merge preserving existing data
                self.metrics_df = empty_df.combine_first(loaded_data)
            else:
                print("Warning: Invalid data format in DB file. Initializing new DB.")
                self.metrics_df = self._initialize_empty_db()
                
        except Exception as e:
            print(f"Error loading DB: {str(e)}. Initializing new DB.")
            self.metrics_df = self._initialize_empty_db()

    def _clean_db(self) -> None:
        """Ensure data integrity through cleaning operations"""
        if not self.metrics_df.empty:
            # Remove duplicates
            self.metrics_df = self.metrics_df.drop_duplicates(subset=['ID'], keep='last')
            
            # Clean data types
            numeric_cols = self.metrics_df.select_dtypes(include=['number']).columns
            self.metrics_df[numeric_cols] = self.metrics_df[numeric_cols].apply(pd.to_numeric, errors='coerce')
            
            # Remove rows with invalid IDs
            self.metrics_df = self.metrics_df[self.metrics_df['ID'].notna() & 
                                            (self.metrics_df['ID'] != '')]
            
            # Ensure session_date is proper datetime
            if 'session_date' in self.metrics_df.columns:
                self.metrics_df['session_date'] = pd.to_datetime(self.metrics_df['session_date'], errors='coerce')
    

    def update_metrics(self, processed_trial_data) -> bool:
        """
        Safely update metrics with duplicate prevention
        
        Args:
            processed_trial_data: ProcessedTrialData instance
            
        Returns:
            bool: True if update succeeded
        """
        try:
            new_metrics = self.calculate_performance_metrics(processed_trial_data)
            
            if new_metrics.empty:
                print("Warning: No metrics calculated")
                return False
                
            if 'ID' not in new_metrics.columns:
                raise ValueError("Metrics missing ID column")
                
            # Clean new metrics
            new_metrics = new_metrics.drop_duplicates(subset=['ID'])
            new_metrics = new_metrics[new_metrics['ID'].notna()]
            
            if self.metrics_df.empty:
                self.metrics_df = new_metrics.reindex(columns=self.metrics_df.columns)
            else:
                # Update strategy using index alignment
                updated_df = self.metrics_df.set_index('ID').combine_first(
                    new_metrics.set_index('ID')
                ).reset_index()
                
                # Preserve original column order and dtypes
                self.metrics_df = updated_df[self.metrics_df.columns].astype(self.metrics_df.dtypes)
            
            self._save_db()
            return True
            
        except Exception as e:
            print(f"Error updating metrics: {str(e)}")
            return False

    def _save_db(self) -> None:
        """Save with error handling and version compatibility"""
        try:
            with open(self.db_path, 'wb') as f:
                pickle.dump(self.metrics_df, f, protocol=4)
        except Exception as e:
            print(f"Error saving DB: {str(e)}")

    # ======================
    # Data Integrity Methods
    # ======================
    
    def check_for_duplicates(self) -> bool:
        """Check if duplicate IDs exist"""
        return self.metrics_df['ID'].duplicated().any()
    
    def remove_duplicates(self, keep: str = 'last') -> None:
        """
        Remove duplicate entries
        
        Args:
            keep: Which duplicates to keep ('first', 'last', or False to drop all)
        """
        if not self.metrics_df.empty:
            self.metrics_df = self.metrics_df.drop_duplicates(subset=['ID'], keep=keep)
            self._save_db()
    
    def validate_integrity(self) -> Dict:
        """
        Run comprehensive integrity checks
        
        Returns:
            Dict of validation results including:
            - has_duplicates: bool
            - null_ids: int
            - invalid_dates: int
            - numeric_outliers: dict
        """
        results = {
            'has_duplicates': self.check_for_duplicates(),
            'null_ids': self.metrics_df['ID'].isna().sum(),
            'invalid_dates': self.metrics_df['session_date'].isna().sum() if 'session_date' in self.metrics_df.columns else 0,
            'numeric_outliers': self._find_numeric_outliers()
        }
        return results
    
    def _find_numeric_outliers(self) -> Dict:
        """Identify potential numeric outliers using IQR"""
        outliers = {}
        numeric_cols = self.metrics_df.select_dtypes(include=['number']).columns
        
        for col in numeric_cols:
            q1 = self.metrics_df[col].quantile(0.25)
            q3 = self.metrics_df[col].quantile(0.75)
            iqr = q3 - q1
            lower_bound = q1 - (1.5 * iqr)
            upper_bound = q3 + (1.5 * iqr)
            
            outlier_count = ((self.metrics_df[col] < lower_bound) | 
                            (self.metrics_df[col] > upper_bound)).sum()
            if outlier_count > 0:
                outliers[col] = outlier_count
                
        return outliers

    def repair_database(self) -> None:
        """Run comprehensive repair operations"""
        print("Running database repair...")
        original_size = len(self.metrics_df)
        
        # Clean duplicates
        self.remove_duplicates()
        
        # Clean invalid rows
        self.metrics_df = self.metrics_df[self.metrics_df['ID'].notna()]
        
        # Reset dtypes
        self._initialize_empty_db()
        self._save_db()
        
        final_size = len(self.metrics_df)
        print(f"Repair completed. Removed {original_size - final_size} invalid entries.")

    # ======================
    # Display & Export Methods
    # ======================
    
    def display_pretty_table(self, round_decimals=2, highlight_extremes=True,
                             subset_columns=None, max_rows=None, open_in_browser=True):
        """
        Display metrics in a 3-level MultiIndex table: Group (Metadata, Preference, Vis1, Invis, Vis2),
        Condition (Max Constant, Min Constant), Metric (sr, sd, etc).
        """
        import pandas as pd
        import numpy as np
        import tempfile
        import webbrowser
    
        df = self.get_metrics()
        if df.empty:
            print("No data available")
            return None
    
        if 'age_years' in df.columns:
            df = df.sort_values('age_years')
    
        expected_cols = self._initialize_empty_db().columns.tolist()
        df = df[[c for c in df.columns if c in expected_cols]]
    
        if subset_columns:
            subset_columns = [c for c in subset_columns if c in df.columns]
            df = df[subset_columns]
    
        group_map = {'vis1_': 'Vis1', 'invis_': 'Invis', 'vis2_': 'Vis2',
                     'mot_': 'Preference', 'pref_': 'Preference'}
        metadata_cols = ['ID', 'age_years', 'session_date']
    
        def parse_col(c):
            if c in metadata_cols:
                return ('Metadata', '', c)
            for pfx, grp in group_map.items():
                if c.startswith(pfx):
                    rest = c[len(pfx):]
                    if rest.endswith('_max_const'):
                        return (grp, 'Max Constant', rest[:-10])
                    if rest.endswith('_min_const'):
                        return (grp, 'Min Constant', rest[:-10])
                    return (grp, '', rest)
            return ('Other', '', c)
    
        df.columns = pd.MultiIndex.from_tuples([parse_col(c) for c in df.columns])
    
        order = ['Metadata', 'Preference', 'Vis1', 'Invis', 'Vis2', 'Other']
        df = df.reindex(
            columns=pd.MultiIndex.from_tuples(
                sorted(df.columns, key=lambda x: (order.index(x[0]) if x[0] in order else 99, x[1], x[2]))
            )
        )
        df = df.round(round_decimals)
    
        if max_rows is not None:
            df = df.head(max_rows)
    
        fmt = lambda x: "" if pd.isna(x) else (f"{x:.{round_decimals}f}" if np.issubdtype(type(x), np.number) else str(x))
        styled = df.style.format(fmt)
    
        if highlight_extremes and hasattr(self, '_highlight_extremes'):
            styled = styled.apply(self._highlight_extremes, axis=0)
    
        sr_cols = [c for c in df.columns if c[2] == 'sr']
        if sr_cols:
            styled = styled.background_gradient(subset=sr_cols, cmap='RdYlGn', vmin=0, vmax=1)
    
        styled = styled.set_table_styles([{
            'selector': 'th',
            'props': [('background-color', '#40466e'), ('color', 'white'), ('font-weight', 'bold')]
        }]).set_properties(border='1px solid #ddd', text_align='center', font_family='Arial')
    
        if ('Metadata', '', 'age_years') in df.columns:
            styled = styled.format({('Metadata', '', 'age_years'): '{:.1f} yrs'})
        if ('Metadata', '', 'session_date') in df.columns:
            styled = styled.format({('Metadata', '', 'session_date'): lambda x: x.strftime('%Y-%m-%d') if pd.notnull(x) else ''})
    
        if open_in_browser:
            with tempfile.NamedTemporaryFile('w', delete=False, suffix='.html') as f:
                f.write(styled.to_html())
                webbrowser.open(f.name)
                print(f"Table opened in browser: {f.name}")
    
        return styled

    def _highlight_extremes(self, s: pd.Series) -> List[str]:
        """Helper for highlighting min/max values"""
        styles = [''] * len(s)
        if s.dtype.kind in 'fi':  # Numeric types
            styles[s.argmin()] = 'background-color: #ffcccc'  # Light red
            styles[s.argmax()] = 'background-color: #ccffcc'   # Light green
        return styles

    def export_to_excel(self, file_path: str, 
                       include_styled: bool = True,
                       sheet_name: str = 'Metrics') -> None:
        """
        Export metrics to Excel with optional styling
        
        Args:
            file_path: Output file path
            include_styled: Whether to include formatted version
            sheet_name: Base sheet name
        """
        try:
            with pd.ExcelWriter(file_path, engine='openpyxl') as writer:
                # Raw data
                self.metrics_df.to_excel(writer, sheet_name=f"{sheet_name}_Raw", index=False)
                
                # Styled version
                if include_styled:
                    styled = self.display_pretty_table()
                    if styled is not None:
                        styled.to_excel(writer, sheet_name=sheet_name, index=False)
                        
        except Exception as e:
            print(f"Error exporting to Excel: {str(e)}")

    # ======================
    # Data Access Methods
    # ======================
    
    def get_metrics(self, 
                   subject_id: Optional[Union[str, List[str]]] = None,
                   sort_by: Optional[str] = None, 
                   ascending: bool = True) -> pd.DataFrame:
        """
        Get metrics with optional filtering
        
        Args:
            subject_id: Single ID or list of IDs to filter
            sort_by: Column to sort by
            ascending: Sort order
            
        Returns:
            Filtered DataFrame
        """
        df = self.metrics_df.copy()
        
        if subject_id is not None:
            if isinstance(subject_id, str):
                df = df[df['ID'] == subject_id]
            else:
                df = df[df['ID'].isin(subject_id)]
                
        if sort_by and sort_by in df.columns:
            df = df.sort_values(sort_by, ascending=ascending)
            
        return df

    def filter_metrics(self, conditions: Dict) -> pd.DataFrame:
        """
        Advanced metric filtering
        
        Args:
            conditions: {
                'column1': value,               # Exact match
                'column2': (min_val, max_val),  # Range
                'column3': [val1, val2]         # In list
            }
            
        Returns:
            Filtered DataFrame
        """
        df = self.metrics_df.copy()
        
        for col, condition in conditions.items():
            if col not in df.columns:
                continue
                
            if isinstance(condition, (tuple, list)):
                if len(condition) == 2:  # Range query
                    min_val, max_val = condition
                    if min_val is not None:
                        df = df[df[col] >= min_val]
                    if max_val is not None:
                        df = df[df[col] <= max_val]
                else:  # List of values
                    df = df[df[col].isin(condition)]
            else:  # Exact match
                df = df[df[col] == condition]
        
        return df

    def get_subject_ids(self) -> List[str]:
        """Get list of all unique subject IDs"""
        return self.metrics_df['ID'].dropna().unique().tolist()

    # ======================
    # Core Metrics Calculation
    # ======================
    
    def calculate_performance_metrics(self, processed_trial_data) -> pd.DataFrame:
        """
        Comprehensive performance metrics calculation
        
        Args:
            processed_trial_data: ProcessedTrialData instance
            
        Returns:
            DataFrame with all calculated metrics
        """
        results = []
        
        def calculate_avg_strides_between_successes(df: pd.DataFrame) -> Optional[float]:
            """Helper to calculate strides between successes"""
            if df is None or 'Success' not in df.columns:
                return None
                
            df = df.reset_index(drop=True)
            success_positions = df.index[df['Success'] == 1].tolist()
            
            if len(success_positions) < 2:
                return None
                
            return np.mean(np.diff(success_positions))
        
        def get_period_data(df: pd.DataFrame, constant_condition: str, length: int = 20):
            """Get data for specific condition period"""
            if df is None or df.empty:
                return None, None
                
            min_target = df['Target size'].min()
            target_tolerance = 0.001
            
            min_target_periods = df[df['Target size'] <= min_target + target_tolerance]
            if min_target_periods.empty:
                return None, None
                
            const_value = (
                min_target_periods['Constant'].max() if constant_condition == 'max'
                else min_target_periods['Constant'].min()
            )
            
            period_data = min_target_periods[
                np.isclose(min_target_periods['Constant'], const_value, rtol=1e-5)
            ]
            
            if period_data.empty:
                return None, None
                
            return period_data.tail(length), period_data.index

        for subject_id, data in processed_trial_data.processed_data.items():
            subject_result = {
                'ID': subject_id,
                'age_years': data['metadata'].get('age_months', float('nan')) / 12,
                'session_date': data['metadata'].get('Session Date')
            }
            
            for trial_type in ['vis1', 'invis', 'vis2']:
                trial_dict = data['trial_data'].get(trial_type)
                df = processed_trial_data.get_trial_df(trial_dict)
                
                if df is None or 'Success' not in df.columns:
                    continue
                
                # Calculate metrics for max constant condition
                max_data, max_indices = get_period_data(df, 'max')
                if max_data is not None:
                    subject_result.update({
                        f'{trial_type}_sr_max_const': max_data['Success'].mean(),
                        f'{trial_type}_sd_max_const': max_data['Sum of gains and steps'].std(),
                        f'{trial_type}_msl_max_const': max_data['Sum of gains and steps'].mean(),
                        f'{trial_type}_error_max_const': (max_data['Sum of gains and steps'] - max_data['Constant']).mean(),
                        f'{trial_type}_asymmetry_max_const': (
                            ((max_data['Right step length'] - max_data['Left step length']) / 
                            (max_data['Right step length'] + max_data['Left step length'])).abs().mean() 
                            if all(col in max_data.columns for col in ['Right step length', 'Left step length']) 
                            else None
                        ),
                        f'{trial_type}_strides_between_success_max_const': calculate_avg_strides_between_successes(max_data),
                        f'{trial_type}_max_const_indices': max_indices
                    })
                
                # Calculate metrics for min constant condition
                min_data, min_indices = get_period_data(df, 'min')
                if min_data is not None:
                    subject_result.update({
                        f'{trial_type}_sr_min_const': min_data['Success'].mean(),
                        f'{trial_type}_sd_min_const': min_data['Sum of gains and steps'].std(),
                        f'{trial_type}_msl_min_const': min_data['Sum of gains and steps'].mean(),
                        f'{trial_type}_error_min_const': (min_data['Sum of gains and steps'] - min_data['Constant']).mean(),
                        f'{trial_type}_asymmetry_min_const': (
                            ((min_data['Right step length'] - min_data['Left step length']) / 
                            (min_data['Right step length'] + min_data['Left step length'])).abs().mean()
                            if all(col in min_data.columns for col in ['Right step length', 'Left step length'])
                            else None
                        ),
                        f'{trial_type}_strides_between_success_min_const': calculate_avg_strides_between_successes(min_data),
                        f'{trial_type}_min_const_indices': min_indices
                    })
                
                # Add trial metadata
                subject_result.update({
                    f'{trial_type}_min_target_size': df['Target size'].min() if 'Target size' in df.columns else None,
                    f'{trial_type}_max_constant': df['Constant'].max() if 'Constant' in df.columns else None,
                    f'{trial_type}_min_constant': df['Constant'].min() if 'Constant' in df.columns else None
                })
                
                # Determine condition order for invis trials
                if trial_type == 'invis':
                    all_max_indices = df.index[df['Constant'] == df['Constant'].max()].tolist()
                    all_min_indices = df.index[df['Constant'] == df['Constant'].min()].tolist()
                    
                    if all_max_indices and all_min_indices:
                        first_max = min(all_max_indices)
                        first_min = min(all_min_indices)
                        subject_result.update({
                            'invis_max_first': first_max < first_min,
                            'invis_min_first': first_min < first_max
                        })
            
            # Process preference trial data
            pref_trial_dict = data['trial_data'].get('pref')
            if pref_trial_dict and isinstance(pref_trial_dict, dict):
                pref_df = pref_trial_dict.get('data')
                if pref_df is not None and not pref_df.empty:
                    # Check for required columns
                    has_right_step = 'Right step length' in pref_df.columns
                    has_left_step = 'Left step length' in pref_df.columns
                    
                    if has_right_step and has_left_step:
                        # Get non-zero steps
                        right_steps = pref_df['Right step length'][pref_df['Right step length'] != 0].dropna()
                        left_steps = pref_df['Left step length'][pref_df['Left step length'] != 0].dropna()
                        
                        if not right_steps.empty and not left_steps.empty:
                            # Calculate motor noise
                            final_right = right_steps.iloc[-1]
                            final_left = left_steps.iloc[-1]
                            
                            norm_right = right_steps / final_right
                            norm_left = left_steps / final_left
                            
                            min_length = min(len(norm_right), len(norm_left))
                            sum_steps = norm_right.iloc[:min_length] + norm_left.iloc[:min_length]
                            subject_result['mot_noise'] = sum_steps.tail(20).std()
                            
                            # Calculate step length asymmetry
                            last_20_right = right_steps.tail(20) / final_right
                            last_20_left = left_steps.tail(20) / final_left
                            
                            asymmetry = np.abs(
                                (last_20_right.values - last_20_left.values) / 
                                (last_20_right.values + last_20_left.values)
                            )
                            subject_result['pref_asymmetry'] = np.mean(asymmetry)
            
            results.append(subject_result)
        
        return pd.DataFrame(results).infer_objects()

    def perform_binary_classification(self, trial_type='invis', condition='max_const', 
                                    model_type='random_forest', test_size=0.2, 
                                    random_state=42, threshold=0.68):
        """
        Perform binary classification to predict if success rate is above or below threshold (default 68%).
        
        Parameters:
        -----------
        trial_type : str ('vis1', 'invis', 'vis2')
            Which trial type to analyze
        condition : str ('max_const', 'min_const')
            Which condition to analyze
        model_type : str ('random_forest' or 'logistic')
            Type of classifier to use
        test_size : float
            Proportion of data to use for testing
        random_state : int
            Random seed for reproducibility
        threshold : float
            Success rate threshold for binary classification (default 0.68 for 68%)
            
        Returns:
        --------
        dict with classification results including:
            - model: trained classifier
            - y_test: true test labels
            - y_pred: predicted labels
            - y_proba: predicted probabilities (for ROC curve)
            - metrics: dictionary of performance metrics
            - feature_importances: dictionary of feature importances
        """
        # Get performance metrics
        metrics_df = self.get_metrics()
        if 'mot_noise' in metrics_df.columns:
            metrics_df = metrics_df[metrics_df['mot_noise'] <= 0.3]# Changed from calculate_performance_metrics() to get_metrics()
        
        # Define predictors - same as in regression but we'll add some additional features
        predictors = [
            'age_years',
            'mot_noise'
        ]
        
        # Only keep predictors that exist in the dataframe
        predictors = [p for p in predictors if p in metrics_df.columns]
        
        # Target is the success rate converted to binary based on threshold
        target_col = f'{trial_type}_sr_{condition}'
        
        if target_col not in metrics_df.columns:
            raise ValueError(f"Target column {target_col} not found in metrics")
        
        # Prepare data - drop rows with missing values
        valid_data = metrics_df[predictors + [target_col]].dropna()
        
        if len(valid_data) < 10:  # Minimum sample size threshold
            raise ValueError(f"Insufficient data (only {len(valid_data)} valid samples)")
        
        # Create binary target (1 if success rate >= threshold, 0 otherwise)
        y = (valid_data[target_col] >= threshold).astype(int)
        X = valid_data[predictors]
        
        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=test_size, random_state=random_state, stratify=y
        )
        
        # Create and fit model
        if model_type == 'random_forest':
            model = Pipeline([
                ('scaler', StandardScaler()),
                ('classifier', RandomForestClassifier(random_state=random_state))
            ])
        elif model_type == 'logistic':
            model = Pipeline([
                ('scaler', StandardScaler()),
                ('classifier', LogisticRegression(random_state=random_state))
            ])
        else:
            raise ValueError("model_type must be 'random_forest' or 'logistic'")
        
        model.fit(X_train, y_train)
        
        # Make predictions
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:, 1]  # Probabilities for positive class
        
        # Calculate metrics
        metrics = {
            'accuracy': accuracy_score(y_test, y_pred),
            'precision': precision_score(y_test, y_pred),
            'recall': recall_score(y_test, y_pred),
            'f1': f1_score(y_test, y_pred),
            'roc_auc': roc_auc_score(y_test, y_proba),
            'confusion_matrix': confusion_matrix(y_test, y_pred),
            'n_samples': len(valid_data),
            'class_balance': y.mean(),  # Proportion of positive class
            'trial_type': trial_type,
            'condition': condition,
            'threshold': threshold,
            'predictors': predictors
        }
        
        # Get feature importances
        if model_type == 'random_forest':
            importances = model.named_steps['classifier'].feature_importances_
        else:  # logistic regression
            importances = model.named_steps['classifier'].coef_[0]
        
        feature_importances = dict(zip(predictors, importances))
        
        return {
            'model': model,
            'X_test': X_test,
            'y_test': y_test,
            'y_pred': y_pred,
            'y_proba': y_proba,
            'metrics': metrics,
            'feature_importances': feature_importances
        }

    def perform_success_rate_regression(self, 
                                     trial_type='invis', 
                                     condition='max_const',
                                     base_predictors=['age_years', 'mot_noise', 'pref_asymmetry'],
                                     include_all_condition_metrics=False,
                                     selected_condition_metrics=None,
                                     model_type='linear',
                                     test_size=0.2,
                                     random_state=42):
        """
        Enhanced regression with flexible condition-specific predictors.
        
        Parameters:
        -----------
        trial_type : str ('vis1', 'invis', 'vis2')
        condition : str ('max_const' or 'min_const')
        base_predictors : list - predictors not tied to conditions
        include_all_condition_metrics : bool - whether to include all available condition metrics
        selected_condition_metrics : list - specific condition metrics to include
                                     (used only if include_all_condition_metrics=False)
        model_type : str ('linear' or 'random_forest')
        test_size : float
        random_state : int
        
        Returns:
        --------
        dict with regression results
        """
        # Get condition-specific suffix
        suffix = condition
        
        # Available condition-specific metrics
        all_condition_metrics = [
            f'{trial_type}_msl_{suffix}',        # mean stride length
            f'{trial_type}_sd_{suffix}',         # stride length SD
            f'{trial_type}_asymmetry_{suffix}',  # step length asymmetry
            f'{trial_type}_error_{suffix}',      # error from target
            f'{trial_type}_strides_between_success_{suffix}'
        ]
        
        # Get available metrics from the dataframe
        metrics_df = self.get_metrics()
        available_metrics = [m for m in all_condition_metrics 
                            if m in metrics_df.columns]
        
        # Determine which condition metrics to include
        if include_all_condition_metrics:
            condition_metrics = available_metrics
        elif selected_condition_metrics:
            # Validate selected metrics
            condition_metrics = []
            for m in selected_condition_metrics:
                metric_name = f'{trial_type}_{m}_{suffix}' if not m.endswith(suffix) else m
                if metric_name in available_metrics:
                    condition_metrics.append(metric_name)
                else:
                    print(f"Warning: Selected metric {metric_name} not available")
        else:
            condition_metrics = []
        
        # Combine all predictors
        predictors = base_predictors + condition_metrics
        target_col = f'{trial_type}_sr_{suffix}'
        
        # Check if we have enough data
        if len(predictors) < 1:
            raise ValueError("No valid predictors available")
        if target_col not in metrics_df.columns:
            raise ValueError(f"Target column {target_col} not found")
            
        # Rest of the regression implementation...
        valid_data = metrics_df[predictors + [target_col]].dropna()
        
        if len(valid_data) < 10:  # Minimum sample size threshold
            raise ValueError(f"Insufficient data (only {len(valid_data)} valid samples)")
        
        # [Previous regression implementation continues...]
        X = valid_data[predictors]
        y = valid_data[target_col]
        
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=test_size, random_state=random_state
        )
        
        # Model creation and fitting...
        if model_type == 'linear':
            model = Pipeline([
                ('scaler', StandardScaler()),
                ('regressor', LinearRegression())
            ])
        elif model_type == 'random_forest':
            model = RandomForestRegressor(random_state=random_state)
        else:
            raise ValueError("model_type must be 'linear' or 'random_forest'")
        
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        # Return results...
        return {
            'model': model,
            'X_test': X_test,
            'y_test': y_test,
            'y_pred': y_pred,
            'metrics': {
                'r2': r2_score(y_test, y_pred),
                'rmse': np.sqrt(mean_squared_error(y_test, y_pred)),
                'n_samples': len(valid_data),
                'trial_type': trial_type,
                'condition': condition,
                'target': target_col,
                'predictors': predictors
            },
            'feature_importances': (dict(zip(predictors, model.feature_importances_)) 
                                 if model_type == 'random_forest' 
                                 else dict(zip(predictors, np.abs(model.named_steps['regressor'].coef_))))
        }


    def run_success_rate_regression(self, trial_types='all'):
        """
        Runs a regression predicting success rate from trial type, position (categorical),
        and mot_noise, age (continuous) for each participant.
    
        Parameters:
        -----------
        trial_types : 'all' or list of str (e.g., ['invis'])
            Whether to run for all trials or just one (e.g., ['invis'])
    
        Returns:
            Fitted regression model object
        """
        long_rows = []
        valid_trials = ['vis1', 'invis', 'vis2']
        
        if trial_types == 'all':
            trials_to_include = valid_trials
        elif isinstance(trial_types, list):
            trials_to_include = trial_types
        else:
            raise ValueError("trial_types must be 'all' or a list of valid trial strings.")
    
        for trial in trials_to_include:
            for pos in ['max_const', 'min_const']:
                sr_col = f"{trial}_sr_{pos}"
                if sr_col in self.metrics_df.columns:
                    sub_df = self.metrics_df[['ID', sr_col, 'mot_noise', 'age_years']].copy()
                    sub_df = sub_df.rename(columns={sr_col: 'success_rate'})
                    sub_df['trial_type'] = trial
                    sub_df['position'] = pos
                    long_rows.append(sub_df)
    
        if not long_rows:
            print("No success rate data available.")
            return None
    
        df_long = pd.concat(long_rows, ignore_index=True)
        df_long.dropna(subset=['success_rate', 'mot_noise', 'age_years'], inplace=True)
        df_long = df_long[df_long['mot_noise'] <= 0.3]
    
        if isinstance(trial_types, list) and len(trial_types) == 1:
            model = smf.ols('success_rate ~ C(position) + mot_noise + age_years', data=df_long).fit()
        else:
            model = smf.ols('success_rate ~ C(trial_type) + C(position) + mot_noise + age_years', data=df_long).fit()
    
        return model


    def plot_success_rate_predictions(self):
        """
        Plots predicted success rate by trial type and position,
        adjusting for mot_noise and age.
        """
        # Prepare long format dataframe
        long_rows = []
        for trial in ['vis1', 'invis', 'vis2']:
            for pos in ['max_const', 'min_const']:
                sr_col = f"{trial}_sr_{pos}"
                if sr_col in self.metrics_df.columns:
                    sub_df = self.metrics_df[['ID', sr_col, 'mot_noise', 'age_years']].copy()
                    sub_df = sub_df.rename(columns={sr_col: 'success_rate'})
                    sub_df['trial_type'] = trial
                    sub_df['position'] = pos
                    long_rows.append(sub_df)
    
        if not long_rows:
            print("No success rate data available.")
            return
    
        df_long = pd.concat(long_rows, ignore_index=True)
        df_long.dropna(subset=['success_rate', 'mot_noise', 'age_years'], inplace=True)
    
        # Fit the model
        model = smf.ols('success_rate ~ C(trial_type) * C(position) + mot_noise + age_years', data=df_long).fit()
    
        # Create grid of predictors for plotting
        pred_df = pd.DataFrame([
            {'trial_type': t, 'position': p, 'mot_noise': df_long['mot_noise'].mean(), 'age_years': df_long['age_years'].mean()}
            for t in ['vis1', 'invis', 'vis2']
            for p in ['max_const', 'min_const']
        ])
    
        # Predict success rate
        pred_df['predicted_sr'] = model.predict(pred_df)
    
        # Plot
        plt.figure(figsize=(8, 5))
        sns.barplot(
            data=pred_df,
            x='trial_type',
            y='predicted_sr',
            hue='position',
            palette='Set2'
        )
        plt.ylim(0, 1)
        plt.ylabel('Predicted Success Rate')
        plt.title('Predicted Success Rate by Trial Type and Position\n(adjusted for mot_noise and age)')
        plt.legend(title='Position')
        plt.tight_layout()
        plt.show()

    
    def plot_success_rate_regression_betas(self, trial_mode='all', use_abs=False):
        """
        Plots regression coefficients (betas) from success rate regression,
        with options to show for all trials (invis vs vis) or individual trial types.
    
        Parameters:
        -----------
        trial_mode : str ('all', 'vis1', 'invis', 'vis2')
            Whether to run regression across all trials or a specific one.
        use_abs : bool
            Whether to plot absolute beta values (magnitude) or signed.
        """
        model = self.run_success_rate_regression(trial_types=trial_mode)
        if model is None:
            print("No model to plot.")
            return
    
        # Access regression result object (not summary)
        coef_mask = model.model.exog_names != 'Intercept'
        coef_series = model.params.drop('Intercept')
        pvals = model.pvalues.drop('Intercept')
    
        if use_abs:
            coef_plot = coef_series.abs()
        else:
            coef_plot = coef_series
    
        # Human-friendly label mapping
        label_map = {
            'C(trial_type)[T.vis1]': 'Trial: vis1 vs invis',
            'C(trial_type)[T.vis2]': 'Trial: vis2 vs invis',
            'C(position)[T.min_const]': 'Position: min_const vs max_const',
            'mot_noise': 'Motion Noise',
            'age_years': 'Age (years)'
        }
        # Apply label mapping with fallback
        labels = [label_map.get(k, k) for k in coef_plot.index]
    
        fig, ax = plt.subplots(figsize=(8, 5))
        bars = ax.barh(labels, coef_plot.values, color='skyblue', edgecolor='k')
    
        ax.set_xlabel("Beta Coefficient" + (" (absolute)" if use_abs else ""))
        ax.set_title(f"Predictors of Success Rate ({trial_mode})")
    
        # Add text annotations
        for i, (bar, coef, pval) in enumerate(zip(bars, coef_series.values, pvals.values)):
            sign = "" if use_abs else f"{coef:.2f} "
            ax.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height() / 2,
                    f"{sign}(p={pval:.3f})", va='center', fontsize=9)
    
        # Metrics
        r2 = model.rsquared
        fval = model.fvalue
        f_pval = model.f_pvalue
        ax.annotate(f"$R^2$ = {r2:.3f}\nF = {fval:.2f} (p = {f_pval:.3g})",
                    xy=(0.95, 0.05), xycoords='axes fraction',
                    ha='right', va='bottom',
                    fontsize=10, bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray"))
    
        plt.tight_layout()
        filename = f"regression_betas_{trial_mode}_{'abs' if use_abs else 'signed'}.png"
        fig_path = Path("figures") / filename
        fig_path.parent.mkdir(parents=True, exist_ok=True)
        plt.savefig(fig_path, dpi=300)
        plt.show()

    
    # ======================
    # Jupyter Integration
    # ======================
    
    def _repr_html_(self) -> str:
        """Jupyter notebook display integration"""
        styled = self.display_pretty_table(max_rows=10)
        return styled._repr_html_() if styled else "No metrics data available"

```

```python
class Plotter:
    def __init__(self, metrics_db, data_manager = None):
        self.data_manager = data_manager
        self.metrics_db = metrics_db
        self.figures_dir = Path('figures')
        os.makedirs(self.figures_dir, exist_ok=True)
        self.default_style = {
            'scatter': {'alpha': 0.7, 'edgecolors': 'w', 'linewidths': 0.5},
            'line': {'linewidth': 2},
            'grid': {'alpha': 0.3},
            'figsize': (12, 6),
            'cmap': 'viridis'
        }
    
    def _save_figure(self, filename):
        """Helper to save figures with consistent settings"""
        plt.savefig(self.figures_dir / filename, dpi=300)
    
    def _setup_plot(self, nrows=1, ncols=1, figsize=None, **kwargs):
        """Create figure and axes with consistent style"""
        figsize = figsize or self.default_style['figsize']
        fig, axes = plt.subplots(nrows, ncols, figsize=figsize, **kwargs)
        if nrows == 1 and ncols == 1:
            return fig, axes  # axes is a single Axes object
        return fig, axes.ravel()  # Flatten for consistency with multiple subplots
    
    def _add_colorbar(self, fig, mappable, ax=None, label=None, exclude_outliers=True, outlier_threshold=2.0):
        """Add colorbar to figure with optional outlier exclusion
        
        Parameters:
        -----------
        fig : matplotlib.figure.Figure
            The figure to which the colorbar will be added
        mappable : matplotlib.cm.ScalarMappable
            The mappable object (usually from a scatter plot)
        ax : matplotlib.axes.Axes, optional
            The axes to which the colorbar is attached
        label : str, optional
            Label for the colorbar
        exclude_outliers : bool, default=True
            Whether to exclude outliers when setting color limits
        outlier_threshold : float, default=2.0
            Number of standard deviations to consider as outliers
        """
        if ax is None:
            ax = plt.gca()
        
        # Get the data array from the mappable
        data = mappable.get_array()
        
        if exclude_outliers and data is not None and len(data) > 0:
            # Calculate robust limits excluding outliers
            median = np.nanmedian(data)
            mad = 1.4826 * np.nanmedian(np.abs(data - median))  # Median Absolute Deviation
            lower = median - outlier_threshold * mad
            upper = median + outlier_threshold * mad
            
            # Clip the data to these limits for the colorbar
            clipped_data = np.clip(data, lower, upper)
            mappable.set_clim(vmin=np.nanmin(clipped_data), 
                             vmax=np.nanmax(clipped_data))
        
        cbar = fig.colorbar(mappable, ax=ax)
        if label:
            cbar.set_label(label, rotation=270, labelpad=15)
        return cbar
    
    def _plot_trendline(self, ax, x, y):
        """Add trendline with correlation coefficient"""
        import numpy as np
        from scipy.stats import pearsonr
    
        # Convert to numeric and drop missing/invalid entries
        x = pd.to_numeric(x, errors='coerce')
        y = pd.to_numeric(y, errors='coerce')
        valid = x.notna() & y.notna()
        x_clean = x[valid].astype(float)
        y_clean = y[valid].astype(float)
    
        if len(x_clean) < 2:
            return  # Not enough data for trendline
    
        coeffs = np.polyfit(x_clean, y_clean, 1)
        trendline = np.poly1d(coeffs)
        r = pearsonr(x_clean, y_clean)[0]
    
        ax.plot(x_clean, trendline(x_clean), color='black', linestyle='--', linewidth=1)
        ax.annotate(f"r = {r:.2f}", xy=(0.05, 0.95), xycoords='axes fraction',
                    ha='left', va='top', fontsize=10, bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="gray", lw=1))
        
    def plot_metrics(self, trial_type='vis1', metric='sr', 
                     color_by='age_years', show_trendline=True, cmap=None):
        """
        Plot performance metrics with flexible coloring and conditioning
        
        Parameters:
        -----------
        trial_type : str ('vis1', 'invis', 'vis2')
            Which trial type to plot
        metric : str ('success_rate', 'sd', 'msl')
            Which metric to plot
        color_by : str (column name from metrics)
            Variable to use for coloring points
        show_trendline : bool
            Whether to display linear trendline
        cmap : str
            Matplotlib colormap to use
        """
        # Validate inputs
        valid_trial_types = ['vis1', 'invis', 'vis2']
        valid_metrics = ['sr', 'sd', 'msl']
        
        if trial_type not in valid_trial_types:
            raise ValueError(f"Invalid trial type. Must be one of {valid_trial_types}")
        if metric not in valid_metrics:
            raise ValueError(f"Invalid metric. Must be one of {valid_metrics}")
        
        # Get data
        metrics_df = self.metrics_db.get_metrics()
        if metrics_df.empty:
            print("No valid data available for plotting")
            return
        if 'mot_noise' in metrics_df.columns:
            metrics_df = metrics_df[metrics_df['mot_noise'] <= 0.3]
        
        # Set up plot parameters
        ylabels = {
            'sr': 'Success Rate',
            'sd': 'Standard Deviation',
            'msl': 'Mean Stride Length'
        }
        ylims = {
            'sr': (0, 1.05),
            'sd': (0, None),
            'msl': (None, None)
        }
        
        max_col = f'{trial_type}_{metric}_max_const'
        min_col = f'{trial_type}_{metric}_min_const'
        
        # Create side-by-side plots
        fig, (ax1, ax2) = self._setup_plot(ncols=2, figsize=(14, 6))
        
        scatter_handles = []
        for ax, col, title in zip([ax1, ax2], [max_col, min_col], 
                                   ['Max Constant', 'Min Constant']):
            valid_data = metrics_df[metrics_df[col].notna()]
            if valid_data.empty:
                continue
            
            # Scatter plot
            scatter = ax.scatter(
                valid_data['age_years'],
                valid_data[col],
                c=valid_data[color_by] if color_by in metrics_df.columns else 'gray',
                cmap=cmap or self.default_style['cmap'],
                **self.default_style['scatter']
            )
            scatter_handles.append(scatter)
    
            if show_trendline:
                self._plot_trendline(ax, valid_data['age_years'], valid_data[col])
            
            ax.set_title(f'{trial_type.capitalize()}: {title} Condition')
            ax.set_xlabel('Age (years)')
            ax.set_ylabel(ylabels[metric])
            ax.set_ylim(ylims[metric])
            ax.grid(**self.default_style['grid'])
        
        # Add one shared colorbar (if color_by is valid)
        if color_by in metrics_df.columns and scatter_handles:
            sm = ScalarMappable(
                norm=Normalize(vmin=metrics_df[color_by].min(), vmax=metrics_df[color_by].max()),
                cmap=cmap or self.default_style['cmap']
            )
            sm.set_array([])  # Needed for colorbar
            self._add_colorbar(fig, sm, ax=[ax1, ax2], label=color_by.replace('_', ' ').title())

        self._save_figure(f"{trial_type}_{metric}_by_age_{color_by if color_by else 'no_color'}.png")
        plt.show()
    
    def plot_noise_analysis(self, trial_type='vis1', color_by='age_years', 
                          threshold=None, cmap=None, figsize=None, analysis_type='vs_success'):
        """
        Unified noise analysis plotting with multiple visualization options
        
        Parameters:
        -----------
        trial_type : str ('vis1', 'invis', 'vis2')
            Which trial type to analyze
        color_by : str ('age_years' or other column name)
            Variable to use for coloring points
        threshold : float or None
            Optional threshold line for motor noise
        cmap : str
            Matplotlib colormap
        figsize : tuple
            Figure dimensions
        analysis_type : str ('vs_success', 'vs_target', 'flexible')
            Type of analysis to perform
        """
        self.metrics_db.get_metrics()
        
        if metrics_df.empty or 'mot_noise' not in metrics_df.columns:
            print("No motor noise data available")
            return
        
        if analysis_type == 'vs_success':
            self._plot_noise_vs_success(metrics_df, trial_type, color_by, threshold, cmap, figsize)
        elif analysis_type == 'vs_target':
            self._plot_noise_vs_target(metrics_df, trial_type, color_by, threshold, cmap, figsize)
        elif analysis_type == 'flexible':
            self._plot_noise_flexible(metrics_df, trial_type, color_by, threshold, cmap, figsize)
        else:
            raise ValueError("Invalid analysis_type. Must be 'vs_success', 'vs_target', or 'flexible'")
    
    def _plot_noise_vs_success(self, metrics_df, trial_type, color_by, threshold, cmap, figsize):
        """Plot motor noise vs success rate"""
        success_col = f'{trial_type}_sr_max_const'
        valid_data = metrics_df[metrics_df[success_col].notna() & metrics_df['mot_noise'].notna()]
        
        fig, ax = self._setup_plot(figsize=figsize)
        
        # Plotting
        sc = ax.scatter(
            valid_data['mot_noise'],
            valid_data[success_col],
            c=valid_data[color_by] if color_by in metrics_df.columns else None,
            cmap=cmap or self.default_style['cmap'],
            **self.default_style['scatter']
        )
        
        if color_by in metrics_df.columns:
            self._add_colorbar(fig, sc, label=color_by.replace('_', ' ').title())
        
        # Add threshold line if specified
        if threshold is not None:
            ax.axvline(x=threshold, color='r', linestyle='--', 
                      label=f'Noise Threshold ({threshold})')
            ax.axvspan(threshold, valid_data['mot_noise'].max(), 
                      color='red', alpha=0.1)
        
        # Add success rate threshold
        success_threshold = 0.68
        ax.axhline(y=success_threshold, color='g', linestyle=':', 
                  label=f'{success_threshold*100:.0f}% Success')
        
        # Formatting
        ax.set_title(f'Motor Noise vs. Success Rate ({trial_type.capitalize()} Trials)')
        ax.set_xlabel('Motor Noise (SD of Normalized Step Lengths)')
        ax.set_ylabel('Success Rate')
        ax.set_ylim(0, 1.05)
        ax.grid(**self.default_style['grid'])
        ax.legend()
        
        plt.tight_layout()
        self._save_figure(f"noise_vs_success_{trial_type}_{color_by if color_by else 'no_color'}.png")
        plt.show()
    
    def _plot_noise_vs_target(self, metrics_df, trial_type, color_by, threshold, cmap, figsize, target_size=0.3):
        """Plot motor noise vs target threshold"""
        fig, ax = self._setup_plot(figsize=figsize)
        ax.set_ylim(0, 0.60)
        
        # 1 SD threshold (68% of strides must fit within target)
        failure_threshold = target_size / 2
        
        # Color points red if noise exceeds threshold
        colors = np.where(metrics_df['mot_noise'] > failure_threshold, 'red', 'green')
        
        ax.scatter(
            x=metrics_df['age_years'],
            y=metrics_df['mot_noise'],
            c=colors,
            **self.default_style['scatter']
        )
        
        # Add thresholds
        ax.axhline(y=failure_threshold, color='r', linestyle='--', 
                   label=f'1SD Failure Threshold ({failure_threshold:.2f})')
        ax.axhline(y=target_size, color='black', linestyle=':', 
                   label=f'Target Size ({target_size})')
        
        # Shade the "high noise = failure" region
        ax.axhspan(failure_threshold, metrics_df['mot_noise'].max()+1, 
                   color='red', alpha=0.1)
        
        # Labels and title
        ax.set_title('Motor Noise vs. 1SD Failure Threshold (68% Strides Fit)')
        ax.set_xlabel('Age (years)')
        ax.set_ylabel('Motor Noise (SD)')
        ax.legend()
        ax.grid(**self.default_style['grid'])
        
        # Annotate interpretation
        ax.text(0.95, 0.95, 
                'Points above red line: >32% predicted misses',
                transform=ax.transAxes, ha='right', va='top',
                bbox=dict(facecolor='white', alpha=0.8))
        
        plt.tight_layout()
        self._save_figure(f"noise_vs_target_1sd_{target_size}.png")
        plt.show()
    
    def _plot_noise_flexible(self, metrics_df, trial_type, color_by, threshold, cmap, figsize, target_size=0.3):
        """Flexible grid plot of noise vs success for all conditions"""
        fig, axes = self._setup_plot(3, 2, figsize=(15, 10) if figsize is None else figsize)
        trial_types = ['vis1', 'invis', 'vis2']
        conditions = ['max_const', 'min_const']
        
        # Calculate thresholds
        threshold_1sd = target_size / 2
        success_ref = 0.68
        
        # Prepare coloring
        if color_by == 'age':
            norm = Normalize(
                vmin=metrics_df['age_years'].min(), 
                vmax=metrics_df['age_years'].max()
            )
            cmap = cmap or self.default_style['cmap']
        else:
            cmap = None
        
        scatter_plots = []
        
        for i, trial_type in enumerate(trial_types):
            for j, condition in enumerate(conditions):
                ax = axes[i*2 + j]
                
                # Get success rate column
                sr_col = f'{trial_type}_sr_{condition}'
                if sr_col not in metrics_df.columns:
                    ax.axis('off')
                    continue
                
                # Filter valid data
                valid_data = metrics_df.dropna(subset=['mot_noise', sr_col])
                if valid_data.empty:
                    ax.axis('off')
                    continue
                
                # Set point colors
                if color_by == 'age':
                    colors = valid_data['age_years']
                else:
                    colors = np.where(
                        valid_data['mot_noise'] > threshold_1sd,
                        'red', 'green'
                    )
                
                # Plot
                sc = ax.scatter(
                    valid_data['mot_noise'],
                    valid_data[sr_col],
                    c=colors,
                    cmap=cmap,
                    norm=norm if color_by == 'age' else None,
                    **self.default_style['scatter']
                )
                
                if color_by == 'age':
                    scatter_plots.append(sc)
                
                # Add threshold shadings
                ax.axvspan(threshold_1sd, valid_data['mot_noise'].max()+1, 
                           color='red', alpha=0.1)
                ax.axhspan(success_ref, 1.05, 0, threshold_1sd/0.4,
                          color='green', alpha=0.1)
                
                # Add reference lines
                ax.axvline(threshold_1sd, color='black', linestyle='--', 
                          label=f'1SD Threshold ({threshold_1sd:.2f})')
                ax.axhline(success_ref, color='gray', linestyle=':', 
                          label='68% Success Reference')
                
                # Formatting
                ax.set_title(f'{trial_type.capitalize()} ({condition.replace("_", " ")})')
                ax.set_xlabel('Motor Noise (SD)')
                ax.set_ylabel('Success Rate')
                ax.set_xlim(0, 0.4)
                ax.set_ylim(0, 1.05)
                ax.grid(**self.default_style['grid'])
                ax.legend(loc='upper right', fontsize=8)
        
        # Add colorbar (only for age coloring)
        if color_by == 'age' and scatter_plots:
            fig.colorbar(
                scatter_plots[0],
                ax=axes.ravel().tolist(),
                shrink=0.6,
                label='Age (years)'
            )
        
        plt.suptitle(
            f'Motor Noise vs. Success Rate\n(1SD Threshold = {threshold_1sd:.2f}, Target Size = {target_size})',
            y=1.02,
            fontsize=14
        )
        self._save_figure(f"noise_vs_success_flexible_{color_by}.png")
        plt.show()
    
    def plot_stride_analysis(self, trial_type='vis1', stride_col='Sum of gains and steps', 
                           figsize=None, age_bins=None, cmap=None, use_rolling_avg=False, 
                           rolling_window=5):
        """
        Unified stride analysis plotting
        
        Parameters:
        -----------
        trial_type : str
            Type of trial to plot ('vis1', 'invis', or 'vis2')
        stride_col : str
            Column name containing stride length data
        figsize : tuple
            Figure size for single plots
        age_bins : list or None
            Bins to group ages
        cmap : str
            Colormap name for age groups
        use_rolling_avg : bool
            Whether to apply rolling average smoothing
        rolling_window : int
            Window size for rolling average
        """
        if not self.data_manager:
            raise ValueError("Data manager required for stride analysis plots")
        if trial_type not in ['vis1', 'invis', 'vis2']:
            raise ValueError("Invalid trial type. Must be 'vis1', 'invis', or 'vis2'")
            
        self.metrics_db.get_metrics()
        
        # Data collection with proper separation
        max_first_trials = []
        min_first_trials = []
        other_trials = []
        
        for subject_id, data in self.data_manager.processed_data.items():
            df = self.data_manager.get_trial_df(data['trial_data'].get(trial_type))
            if df is None or df.empty or stride_col not in df.columns:
                continue
                
            age = data['metadata'].get('age_years')
            if age is None:
                continue
                
            # Filter out stride number 0 if present
            df = df[df['Stride Number'] != 0].copy() if 'Stride Number' in df.columns else df.copy()
            
            # For trial type: use the pre-calculated max_first/min_first flags
            if trial_type == 'invis':
                subject_metrics = metrics_df[metrics_df['ID'] == subject_id]
                if not subject_metrics.empty:
                    subject_metrics = subject_metrics.iloc[0]
                    if subject_metrics[f'{trial_type}_max_first']:
                        max_first_trials.append((df, age))
                    elif subject_metrics[f'{trial_type}_min_first']:
                        min_first_trials.append((df, age))
                    else:
                        other_trials.append((df, age))
                else:
                    other_trials.append((df, age))
            else:
                # For non-trial types, just collect all data
                max_first_trials.append((df, age))
        
        # Create plots based on data availability
        if trial_type == 'invis' and (max_first_trials or min_first_trials):
            fig, axes = self._setup_plot(1, 2, figsize=(20, 8) if figsize is None else figsize)
            ax_max, ax_min = axes
            
            # Plot max-first trials if any exist
            if max_first_trials:
                try:
                    max_data = pd.concat([df.assign(Age=age) for df, age in max_first_trials])
                    self._plot_stride_group(
                        ax_max, max_data, stride_col, 
                        "Trials with Maximum Target First", 
                        age_bins, cmap or self.default_style['cmap'], 
                        use_rolling_avg, rolling_window
                    )
                except ValueError as e:
                    print(f"Could not plot max-first trials: {str(e)}")
                    ax_max.text(0.5, 0.5, 'No max-first trial data', 
                               ha='center', va='center')
                    ax_max.set_title("No max-first trial data")
            
            # Plot min-first trials if any exist
            if min_first_trials:
                try:
                    min_data = pd.concat([df.assign(Age=age) for df, age in min_first_trials])
                    self._plot_stride_group(
                        ax_min, min_data, stride_col, 
                        "Trials with Minimum Target First", 
                        age_bins, cmap or self.default_style['cmap'], 
                        use_rolling_avg, rolling_window
                    )
                except ValueError as e:
                    print(f"Could not plot min-first trials: {str(e)}")
                    ax_min.text(0.5, 0.5, 'No min-first trial data', 
                               ha='center', va='center')
                    ax_min.set_title("No min-first trial data")
            
            plt.tight_layout()
            self._save_figure(f"stride_length_by_age_{trial_type}_{'rolling' if use_rolling_avg else 'raw'}.png")
            plt.show()
        else:
            # Single plot for non-trial types or if only one group exists
            combined_data = max_first_trials + min_first_trials + other_trials
            
            if combined_data:
                try:
                    combined_df = pd.concat([df.assign(Age=age) for df, age in combined_data])
                    fig, ax = self._setup_plot(figsize=figsize)
                    self._plot_stride_group(
                        ax, combined_df, stride_col, 
                        trial_type.capitalize(), 
                        age_bins, cmap or self.default_style['cmap'], 
                        use_rolling_avg, rolling_window
                    )
                    plt.tight_layout()
                    self._save_figure(f"stride_length_by_age_{trial_type}_{'rolling' if use_rolling_avg else 'raw'}.png")
                    plt.show()
                except ValueError as e:
                    print(f"Could not create plot: {str(e)}")
            else:
                print("No valid data available for plotting")
    
    def _plot_stride_group(self, ax, data, stride_col, title, age_bins, cmap, use_rolling_avg, rolling_window):
        """Helper function to plot one group of strides"""
        # Age grouping
        if age_bins is None:
            data['Age Group'] = data['Age'].round().astype(int)
            is_binned = False
        else:
            data['Age Group'] = pd.cut(data['Age'], bins=age_bins)
            is_binned = True
        
        # Calculate averages
        target_params = data.groupby('Stride Number').agg({
            'Constant': 'mean',
            'Target size': 'mean'
        }).reset_index()
        
        stride_means = data.groupby(['Age Group', 'Stride Number']).agg({
            stride_col: 'mean'
        }).reset_index()
        
        # Get colors
        age_groups = sorted(stride_means['Age Group'].unique())
        colormap = plt.colormaps.get_cmap(cmap)
        colors = colormap(np.linspace(0, 1, len(age_groups)))
        
        # Plot target regions
        for _, row in target_params.iterrows():
            stride_num = row['Stride Number']
            center = row['Constant']
            height = row['Target size']
            
            ax.add_patch(plt.Rectangle(
                (stride_num - 0.5, center - height/2),
                1, height,
                color='gray', alpha=0.2, zorder=0
            ))
            
            ax.hlines(
                y=center,
                xmin=stride_num - 0.5,
                xmax=stride_num + 0.5,
                colors='black',
                linestyles='--',
                alpha=0.5,
                zorder=1
            )
        
        # Plot stride lines
        for age_group, color in zip(age_groups, colors):
            group_data = stride_means[stride_means['Age Group'] == age_group]
            
            label = (f'{age_group.left:.1f}-{age_group.right:.1f}' 
                    if is_binned else f'{age_group}')
            
            # Apply rolling average if requested
            if use_rolling_avg:
                y_values = (group_data[stride_col]
                           .rolling(window=rolling_window, min_periods=1, center=True)
                           .mean()
                           .values)
            else:
                y_values = group_data[stride_col].values
            
            ax.plot(
                group_data['Stride Number'],
                y_values,
                label=label,
                color=color,
                alpha=0.8,
                linewidth=2,
                zorder=10
            )
        
        # Formatting
        max_stride = target_params['Stride Number'].max()
        ax.set_xticks(np.arange(0, max_stride+1, 25))
        ax.set_ylim(0, 3)
        ax.grid(True, which='major', alpha=0.3)
        ax.set_title(f'{title}\nAverage {stride_col.replace("_", " ")}')
        ax.set_xlabel('Stride Number')
        ax.set_ylabel(stride_col.replace('_', ' '))
        
        # Legend
        ax.plot([], [], color='gray', alpha=0.2, linewidth=10, label='Target Region')
        ax.plot([], [], color='black', linestyle='--', label='Target Center')
        ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    def plot_correlation_analysis(self, x_col, y_col, color_by=None, 
                                show_trendline=True, cmap=None, figsize=None):
        """
        Generalized correlation plot between any two metrics
        
        Parameters:
        -----------
        x_col : str
            Column name for x-axis data
        y_col : str
            Column name for y-axis data
        color_by : str
            Column name for coloring points
        show_trendline : bool
            Whether to show trendline
        cmap : str
            Colormap to use
        figsize : tuple
            Figure size
        """
        self.metrics_db.get_metrics()
        
        if x_col not in metrics_df.columns or y_col not in metrics_df.columns:
            raise ValueError("One or both specified columns not found in data")
            
        valid_data = metrics_df[[x_col, y_col] + ([color_by] if color_by else [])].dropna()
        
        fig, ax = self._setup_plot(figsize=figsize)
        
        # Plot data
        sc = ax.scatter(
            valid_data[x_col],
            valid_data[y_col],
            c=valid_data[color_by] if color_by else None,
            cmap=cmap or self.default_style['cmap'],
            **self.default_style['scatter']
        )
        
        if color_by:
            self._add_colorbar(fig, sc, label=color_by.replace('_', ' ').title())
        
        if show_trendline:
            self._plot_trendline(ax, valid_data[x_col], valid_data[y_col])
        
        # Formatting
        ax.set_xlabel(x_col.replace('_', ' ').title())
        ax.set_ylabel(y_col.replace('_', ' ').title())
        ax.grid(**self.default_style['grid'])
        
        plt.tight_layout()
        self._save_figure(f"correlation_{x_col}_vs_{y_col}.png")
        plt.show()
    
    def plot_regression_results(self, regression_results, figsize=None, save_figure=True):
        """
        Plot the results of regression analysis
        
        Parameters:
        -----------
        regression_results : dict
            Output from regression analysis
        figsize : tuple
            Figure dimensions
        save_figure : bool
            Whether to save the figure
        """
        # Extract data from results
        y_test = regression_results['y_test']
        y_pred = regression_results['y_pred']
        metrics = regression_results['metrics']
        target = metrics['target']
        
        # Create figure
        fig, ax = self._setup_plot(figsize=figsize)
        
        # Plot actual vs predicted
        ax.scatter(y_test, y_pred, alpha=0.7, edgecolors='w', linewidths=0.5)
        
        # Add perfect prediction line
        lims = [
            min(ax.get_xlim()[0], ax.get_ylim()[0]),
            max(ax.get_xlim()[1], ax.get_ylim()[1]),
        ]
        ax.plot(lims, lims, 'k--', alpha=0.75, zorder=0)
        
        # Add text with metrics
        textstr = '\n'.join((
            f'R¬≤ = {metrics["r2"]:.2f}',
            f'RMSE = {metrics["rmse"]:.2f}',
            f'n = {metrics["n_samples"]}'
        ))
        props = dict(boxstyle='round', facecolor='white', alpha=0.5)
        ax.text(0.05, 0.95, textstr, transform=ax.transAxes,
                fontsize=12, verticalalignment='top', bbox=props)
        
        # Labels and title
        ax.set_xlabel(f'Actual {target}')
        ax.set_ylabel(f'Predicted {target}')
        ax.set_title('Regression Model Performance')
        ax.grid(**self.default_style['grid'])
        
        plt.tight_layout()
        
        if save_figure:
            filename = f"regression_{target}_{'_'.join(metrics['predictors'])}.png"
            self._save_figure(filename)
        
        plt.show()
        
        # Plot feature importances if available
        if regression_results.get('feature_importances'):
            fig, ax = self._setup_plot(figsize=(10, 4))
            features = list(regression_results['feature_importances'].keys())
            importances = list(regression_results['feature_importances'].values())
            
            # Sort by importance
            sorted_idx = np.argsort(importances)[::-1]
            features = [features[i] for i in sorted_idx]
            importances = [importances[i] for i in sorted_idx]
            
            ax.bar(features, importances)
            ax.set_title('Feature Importances')
            ax.set_ylabel('Importance')
            plt.xticks(rotation=45)
            plt.tight_layout()
            
            if save_figure:
                filename = f"feature_importance_{target}_{'_'.join(metrics['predictors'])}.png"
                self._save_figure(filename)
            
            plt.show()


    def plot_all_regression_condition_combinations(self, selected_condition_metrics, model_type='linear', save_figure=True):
        """
        Plot regression results for all combinations of trial types and conditions,
        using only the specified condition metrics.
        
        Parameters:
        -----------
        selected_condition_metrics : list
            List of condition metrics to include (without trial_type or condition suffix)
            e.g., ['msl', 'sd', 'error']
        model_type : str
            Type of regression model ('linear' or 'random_forest')
        save_figure : bool
            Whether to save figures to file
        """
        # Define all possible combinations
        trial_types = ['vis1', 'invis', 'vis2']
        conditions = ['max_const', 'min_const']
        
        # Base predictors (common to all regressions)
        base_predictors = ['age_years', 'mot_noise', 'pref_asymmetry']
        
        # Iterate through all combinations
        for trial_type in trial_types:
            for condition in conditions:
                try:
                    # Perform regression with selected metrics only
                    results = self.metrics_db.perform_success_rate_regression(
                        trial_type=trial_type,
                        condition=condition,
                        base_predictors=base_predictors,
                        include_all_condition_metrics=False,
                        selected_condition_metrics=selected_condition_metrics,
                        model_type=model_type
                    )
                    
                    # Plot results
                    self.plot_regression_results(results, save_figure=save_figure)
                    
                except ValueError as e:
                    print(f"Skipping {trial_type}-{condition}: {str(e)}")
                    continue


    
    def plot_classification_results(self, classification_results, figsize=(15, 5), save_figure=True):
        """
        Plot the results of the binary classification.
        
        Parameters:
        -----------
        classification_results : dict
            Output from perform_binary_classification()
        figsize : tuple
            Figure dimensions
        save_figure : bool
            Whether to save the figure to file
        """
        # Extract data from results
        metrics = classification_results['metrics']
        y_test = classification_results['y_test']
        y_pred = classification_results['y_pred']
        y_proba = classification_results['y_proba']
        feature_importances = classification_results['feature_importances']
        
        # Create figure with subplots
        fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=figsize)
        
        # Plot 1: Confusion matrix
        cm = metrics['confusion_matrix']
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax1,
                   xticklabels=['Below', 'Above'], 
                   yticklabels=['Below', 'Above'])
        ax1.set_title('Confusion Matrix')
        ax1.set_xlabel('Predicted')
        ax1.set_ylabel('Actual')
        
        # Plot 2: ROC curve
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        ax2.plot(fpr, tpr, label=f'AUC = {metrics["roc_auc"]:.2f}')
        ax2.plot([0, 1], [0, 1], 'k--')
        ax2.set_xlabel('False Positive Rate')
        ax2.set_ylabel('True Positive Rate')
        ax2.set_title('ROC Curve')
        ax2.legend(loc='lower right')
        ax2.grid(True, alpha=0.3)
        
        # Plot 3: Feature importances
        features = list(feature_importances.keys())
        importances = list(feature_importances.values())
        
        # Sort by importance
        sorted_idx = np.argsort(importances)[::-1]
        features = [features[i] for i in sorted_idx]
        importances = [importances[i] for i in sorted_idx]
        
        ax3.barh(features, importances)
        ax3.set_title('Feature Importances')
        ax3.set_xlabel('Importance')
        ax3.invert_yaxis()  # Most important at top
        
        # Add text with metrics
        textstr = '\n'.join((
            f'Accuracy: {metrics["accuracy"]:.2f}',
            f'Precision: {metrics["precision"]:.2f}',
            f'Recall: {metrics["recall"]:.2f}',
            f'F1: {metrics["f1"]:.2f}',
            f'AUC: {metrics["roc_auc"]:.2f}',
            f'n: {metrics["n_samples"]}',
            f'Threshold: {metrics["threshold"]*100:.0f}%'
        ))
        props = dict(boxstyle='round', facecolor='white', alpha=0.5)
        ax3.text(1.05, 0.95, textstr, transform=ax3.transAxes,
                fontsize=10, verticalalignment='top', bbox=props)
        
        plt.tight_layout()
        
        if save_figure:
            filename = (f"classification_{metrics['trial_type']}_{metrics['condition']}_"
                       f"{metrics['threshold']*100:.0f}percent.png")
            plt.savefig(self.figures_dir / filename, dpi=300, bbox_inches='tight')
        
        plt.show()

    def plot_sl_change_after_success(self, trial_type='invis', age_bins=[7, 10, 13, 17], use_entire_period=False):
        """
        Plot the change in 'Sum of gains and steps' after a success vs after a not success,
        separated by max and min constant, and stacked by age groups.
    
        Parameters:
        -----------
        trial_type : str
            Which trial type to use ('vis1', 'invis', 'vis2')
        age_bins : list
            List of age bin edges (default: [7, 10, 13, 17])
        use_entire_period : bool
            If True, use entire period where constant == max or min, 
            otherwise use periods with min target size (default: False)
        """
        metrics_df = self.metrics_db.get_metrics()
        if metrics_df.empty:
            print("No data available.")
            return
        
        # Prepare age labels
        age_labels = [f'{age_bins[i]}-{age_bins[i+1]}' for i in range(len(age_bins)-1)]
        metrics_df['age_group'] = pd.cut(metrics_df['age_years'], bins=age_bins, labels=age_labels)
        
        # Initialize subplots: one column for Max, one for Min, rows for each age group
        fig, axes = plt.subplots(len(age_labels), 2, figsize=(12, 4 * len(age_labels)), sharey=True)
        
        if len(age_labels) == 1:
            axes = np.expand_dims(axes, axis=0)  # Ensure consistent indexing
    
        for i, age_group in enumerate(age_labels):
            # Select subjects in this age group
            group_subjects = metrics_df[metrics_df['age_group'] == age_group]
            
            max_data = []
            min_data = []
            
            for _, row in group_subjects.iterrows():
                subject_id = row['ID']
                trial_dict = self.data_manager.processed_data[subject_id]['trial_data'].get(trial_type)
                data = self.data_manager.get_trial_df(trial_dict)
                if data is None:
                    continue
    
                for const_type, ax in zip(['max', 'min'], axes[i]):
                    if use_entire_period:
                        # Entire period when Constant == max or min
                        if const_type == 'max':
                            const_value = data['Constant'].max()
                        else:
                            const_value = data['Constant'].min()
                        
                        relevant_data = data[np.isclose(data['Constant'], const_value, rtol=1e-5)]
                    else:
                        # Use period identified by _get_period_data
                        indices = row[f'{trial_type}_{const_type}_const_indices']
                        if indices is not None:
                            relevant_data = data.loc[indices]
                        else:
                            relevant_data = None
    
                    if relevant_data is None or relevant_data.empty:
                        ax.set_visible(False)
                        continue
    
                    # Calculate the delta Sum of gains and steps after each stride
                    relevant_data = relevant_data.sort_values('Stride Number')
                    relevant_data['Delta'] = relevant_data['Sum of gains and steps'].diff().shift(-1)
    
                    # Categorize by Success
                    success_deltas = relevant_data[relevant_data['Success'] == 1]['Delta'].dropna()
                    not_success_deltas = relevant_data[relevant_data['Success'] == 0]['Delta'].dropna()
                    
                    if const_type == 'max':
                        max_data.append((success_deltas, not_success_deltas))
                    else:
                        min_data.append((success_deltas, not_success_deltas))
    
            for cond_data, ax, title in zip([max_data, min_data], axes[i], ['Max Constant', 'Min Constant']):
                # Flatten the data
                success_vals = [val for subj in cond_data for val in subj[0]]
                not_success_vals = [val for subj in cond_data for val in subj[1]]
                
                if success_vals and not_success_vals:
                    sns.violinplot(
                        data=[success_vals, not_success_vals],
                        palette=['green', 'red'],
                        ax=ax,
                        cut=0
                    )
                    ax.set_xticks([0, 1])
                    ax.set_xticklabels(['Success', 'Failure'])
                    ax.set_title(f'{title} ({age_group} yrs)')
                    ax.set_ylabel('Œî Sum of Gains and Steps')
                    ax.set_ylim(-1, 1)  # Restrict the y-axis range to -1, 1
                    ax.grid(True, alpha=0.3)
                else:
                    ax.set_visible(False)
        
        plt.tight_layout(rect=[0, 0, 1, 0.94])  # more space for title
        period_text = "Entire Period" if use_entire_period else "Min Target Size Period"
        fig.suptitle(f'Change in Sum of Gains and Steps After Success vs Failure\n'
                     f'Trial Type: {trial_type} | {period_text}', fontsize=16)
        # Save the figure
        filename = f"sum_gain_step_change_{trial_type}_{'entire' if use_entire_period else 'min_target'}.png"
        plt.savefig(self.figures_dir / filename, dpi=300, bbox_inches='tight')
        plt.show()

    def plot_sum_of_gains_steps(self, trial_type='invis', figsize=(14, 8), 
                              cmap='viridis', alpha=0.7, linewidth=1.5,
                              smoothing_window=5):
        """
        Plot each participant's Sum of gains and steps across strides, 
        with target regions and age-based coloring.
        
        Parameters:
        -----------
        trial_type : str ('vis1', 'invis', 'vis2')
            Which trial type to plot
        figsize : tuple
            Figure dimensions
        cmap : str
            Colormap for age gradient
        alpha : float
            Line transparency
        linewidth : float
            Line thickness
        smoothing_window : int
            Number of strides for rolling average smoothing (default: 5)
        """
        if not hasattr(self.data_manager, 'processed_data'):
            print("Error: No processed data available")
            return
        
        # Prepare figure - split into two subplots for invis trials
        if trial_type == 'invis':
            fig, (ax_max_first, ax_min_first) = plt.subplots(1, 2, figsize=(20, 8))
            axes = {'max_first': ax_max_first, 'min_first': ax_min_first}
            ax_max_first.set_title('Higher Target First')
            ax_min_first.set_title('Lower Target First')
        else:
            fig, ax = plt.subplots(figsize=figsize)
            axes = {'all': ax}
        
        # Get age range for colormap
        ages = []
        for subject_data in self.data_manager.processed_data.values():
            ages.append(subject_data['metadata']['age_years'])
        age_min, age_max = min(ages), max(ages)
        norm = plt.Normalize(age_min, age_max)
        cmap = plt.get_cmap(cmap)
        
        # Get metrics to determine target order for invis trials
        metrics_df = self.metrics_db.get_metrics()
        
        # Plot each participant
        for subject_id, subject_data in self.data_manager.processed_data.items():
            trial_data = subject_data['trial_data'].get(trial_type)
            if trial_data is None:
                continue
                
            df = trial_data['data']
            if df is None or 'Sum of gains and steps' not in df.columns:
                continue
                
            age = subject_data['metadata']['age_years']
            color = cmap(norm(age))
            
            # Apply rolling average smoothing
            smoothed_data = df['Sum of gains and steps'].rolling(
                window=smoothing_window,
                min_periods=1,
                center=True
            ).mean()
            
            # Determine which axes to plot on
            if trial_type == 'invis' and metrics_df is not None:
                subject_metrics = metrics_df[metrics_df['ID'] == subject_id]
                if not subject_metrics.empty:
                    if subject_metrics['invis_max_first'].iloc[0]:
                        current_ax = axes['max_first']
                    elif subject_metrics['invis_min_first'].iloc[0]:
                        current_ax = axes['min_first']
                    else:
                        continue  # Skip if order not determined
                else:
                    continue
            else:
                current_ax = axes['all']
            
            # Plot participant's smoothed data
            current_ax.plot(
                df['Stride Number'],
                smoothed_data,
                color=color,
                alpha=alpha,
                linewidth=linewidth
            )
        
        # Plot target regions (using first participant's data as reference)
        for ax in axes.values():
            if len(df) > 0:
                target_center = df['Constant'].mean()
                target_width = df['Target size'].mean()
                
                ax.axhline(
                    y=target_center,
                    color='black',
                    linestyle='--',
                    alpha=0.5,
                    label='Target Center'
                )
                
                ax.fill_between(
                    x=df['Stride Number'],
                    y1=target_center - target_width/2,
                    y2=target_center + target_width/2,
                    color='gray',
                    alpha=0.2,
                    label='Target Region'
                )
            
            # Formatting
            ax.set_xlabel('Stride Number')
            ax.set_ylabel('Sum of Gains and Steps')
            ax.grid(alpha=0.3)
        
        # Main title and colorbar
        fig.suptitle(f'Sum of Gains and Steps by Stride Number ({trial_type.capitalize()} Trials)\n'
                    f'{smoothing_window}-stride moving average', y=1.02)
        
        # Add colorbar
        sm = ScalarMappable(norm=norm, cmap=cmap)
        sm.set_array([])
        if trial_type == 'invis':
            fig.colorbar(sm, ax=list(axes.values()), label='Age (years)')
        else:
            fig.colorbar(sm, ax=axes['all'], label='Age (years)')
        
        # Legend
        legend_elements = [
            Line2D([0], [0], color='black', linestyle='--', label='Target Center'),
            Patch(facecolor='gray', alpha=0.2, label='Target Region')
        ]
        if trial_type == 'invis':
            axes['max_first'].legend(handles=legend_elements)
            axes['min_first'].legend(handles=legend_elements)
        else:
            axes['all'].legend(handles=legend_elements)
        
        # Adjust layout with constrained_layout instead of tight_layout
        fig.set_constrained_layout(True)
        self._save_figure(f"sum_gains_steps_{trial_type}_by_age_smoothed.png")
        plt.show()

```
